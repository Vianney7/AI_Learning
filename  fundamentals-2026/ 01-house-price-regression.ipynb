```python
# Exercise 1: Linear Regression - House Prices
# Author: Atem Vianney | April 28, 2026

import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# 1. Create training data - sqft vs price in thousands
data = pd.DataFrame({
    'size_sqft': [1000, 1200, 1500, 1800, 2000, 2200, 2500, 2800],
    'price_k': [210, 250, 310, 360, 405, 440, 510, 560]
})

print("Training Data:")
print(data)

# 2. Train the AI model - this 1 line is machine learning
model = LinearRegression()
model.fit(data[['size_sqft']], data['price_k'])

# 3. Check what the AI learned
print(f"\nAI found the pattern: For every +1 sqft, price goes up ${model.coef_[0]:.2f}k")
print(f"Base price for 0 sqft: ${model.intercept_:.2f}k")

# 4. Make predictions on new houses
test_houses = [[1300], [1700], [2400]]
predictions = model.predict(test_houses)

for size, price in zip([1300, 1700, 2400], predictions):
    print(f"Predicted price for {size} sqft: ${price:.0f}k")

# 5. Visualize what the AI learned
plt.figure(figsize=(10,6))
plt.scatter(data['size_sqft'], data['price_k'], color='blue', s=100, label='Real house sales')
plt.plot(data['size_sqft'], model.predict(data[['size_sqft']]), color='red', linewidth=3, label='AI model')
plt.scatter([1300, 1700, 2400], predictions, color='green', s=200, marker='X', label='New predictions')
plt.xlabel('House Size (sqft)', fontsize=12)
plt.ylabel('Price ($1000s)', fontsize=12)
plt.title('AI Model: Predicting House Prices from Size', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
